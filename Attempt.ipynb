```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data.csv")
df
```

```python
df['Test'] = df['Design'] + df['Direction'] + df['Displacement Rate'].astype(str)
df.drop(columns=['Design', 'Direction', 'Displacement Rate'])
df['Force'] = pd.to_numeric(df['Force'])
df['Displacement'] = pd.to_numeric(df['Displacement'])
df = df.drop(columns=['Design', 'Direction', 'Displacement Rate'])
df
```

```python
summary = df.groupby('Test').mean().reset_index()
summary
```

```python
fig, ax = plt.subplots()

# Create a bar chart
bars = ax.bar(summary['Test'], summary['Force'], color='blue')

# Add labels
ax.set_xlabel('Test')
ax.set_ylabel('Average Force (kN)')
ax.set_title('Average Force for Each Test')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
